Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

abt_df = pd.read_csv("#2305505.csv")
```

```python
cs = abt_df["CreditScore"]
age = abt_df["Age"]
tnr = abt_df["Tenure"]
bal = abt_df["Balance"]
nop = abt_df["NumOfProducts"]
e_s = abt_df["EstimatedSalary"]

cnt_cs = len(cs)
min_cs = cs.min()
max_cs = cs.max()
avg_cs = cs.mean()
med_cs = cs.median()
q1_cs = cs.quantile(0.25)
q3_cs = cs.quantile(0.75)
iqr_cs = q3_cs - q1_cs
min_acc_cs = q1_cs - 1.5*iqr_cs
max_acc_cs = q3_cs + 1.5*iqr_cs
cs_car = cs.nunique()
cs_miss_per = (cs.isnull().sum()/cnt_cs)
cs_std = cs.std()
cs.hist(bins=60, edgecolor = "Red")   # histogram with 70 bins, bins should be square root of total obervations
plt.title("Credit Score Frequency Distribution", color = "Green")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")
print(cnt_cs)
print(min_cs)
print(max_cs)
print(avg_cs)
print(med_cs)
print(q1_cs)
print(q3_cs)
print(iqr_cs)
print(min_acc_cs)
print(max_acc_cs)
print(cs_car)
print(cs_miss_per)
print(cs_std)
plt.show()
```
